# `pyemsi` in Jupyter Notebook

This notebook is for testing and experimenting with the pyemsi library.

```python
from pyemsi import Plotter
import pyvista as pv
file_path = r"C:\Users\eskan\OneDrive\Documents\Github\pyemsi\.pyemsi\output.pvd"
plt1 = Plotter(notebook=True, backend="static").set_file(file_path)
print(plt1.mesh[0].array_names)
plt1.reader.set_active_time_point(2)
plt1.plotter.view_xy()
plt1.set_scalar("B-Mag (T)", mode="element", cell2point=True, show_edges=True)
plt1.set_vector(name="B-Vec (T)", scale="B-Mag (T)", factor=5e-3)
plt1.show()
```

# Plotting Contours

```python
plt2 = Plotter(notebook=True, backend="static").set_file(file_path)
plt2.plotter.view_xy()
plt2.set_scalar("B-Mag (T)", mode="element", cell2point=True, show_edges=True).set_contour(name="Flux (A/m)").show()
```
